```python
# problem 1
import numpy as np
import matplotlib.pyplot as plt
# Fix the seed value
np.random.seed(0)

# Covariance matrix
cov = [
    [1.0, 0.8],
    [0.8, 1.0]
]
# Generate 500 samples
mean_0 = [-3, 0]
data_0 = np.random.multivariate_normal(mean_0, cov, 500)
mean_1 = [0, -3]
data_1 = np.random.multivariate_normal(mean_1, cov, 500)

# problem 2
plt.figure(figsize=(8, 6))
plt.scatter(data_0[:, 0], data_0[:, 1], alpha=0.6, label ='0', color='blue', edgecolors='k')
plt.scatter(data_1[:, 0], data_1[:, 1], alpha=0.6, label='1', color='red', edgecolors='k')
plt.title("Scatter Plot of two 2D Normal Distribution")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(title="Class")
plt.grid(True)
plt.axis("equal")
plt.show()

# problem 3
# Extract each dimension
x = data_0[:, 0]
y = data_0[:, 1]

# Determine common x-axis range
x_min = min(x.min(), y.min())
x_max = max(x.max(), y.max())

# Plot histograms
plt.figure(figsize=(12, 5))

# Histogram for x
plt.subplot(1, 2, 1)
plt.hist(x, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of X values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.xlim(x_min, x_max)

# Histogram for y
plt.subplot(1, 2, 2)
plt.hist(y, bins=30, color='salmon', edgecolor='black')
plt.title("Histogram of Y values")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.xlim(x_min, x_max)

plt.tight_layout()
plt.show()

# PROBLEM 5
combined_data = np.vstack((data_0, data_1))
print("The shape of the combined data is:",combined_data.shape)

labels_0 = np.zeros((500, 1))  # Label 0 for Problem 1 data
labels_1 = np.ones((500, 1))   # Label 1 for Problem 4 data

# Step 2: Append labels to data
data_with_labels_0 = np.hstack((data_0, labels_0))  # Shape (500, 3)
data_with_labels_1 = np.hstack((data_1, labels_1))  # Shape (500, 3)

# Step 3: Combine both labeled datasets
labeled_data = np.vstack((data_with_labels_0, data_with_labels_1))  # Shape (1000, 3)

# Check result
print("The shape of the final combined data with labelling for first and fourth problem: ",labeled_data.shape) # displays the shape of the final combined data
print(labeled_data[:5]) # displays the first 5 values
```
